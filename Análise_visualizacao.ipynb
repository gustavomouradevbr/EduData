```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Carregamento e Exploração Inicial dos Dados

```python
file_path = '/content/Matrículas_ red_ municipal do_ano_2024.csv'

try:
    df = pd.read_csv(file_path, sep=';', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, sep=';', encoding='latin-1')

print("Primeiras 5 linhas do DataFrame:")
display(df.head())

print("\nInformações do DataFrame:")
df.info()

print("\nEstatísticas descritivas básicas:")
display(df.describe(include='all'))
```

## 1. Exploração Básica

```python
# 1.1 Calcular o total de matrículas.
total_matriculas = len(df)
print(f"Total de Matrículas: {total_matriculas}")

# 1.2 Contar alunos por sexo (Feminino/Masculino).
contagem_sexo = df['SEXO'].value_counts()
print("\nContagem de Alunos por Sexo:")
print(contagem_sexo)

# 1.3 Calcular a idade média, mínima e máxima dos alunos.
idade_media = df['IDADE'].mean()
idade_minima = df['IDADE'].min()
idade_maxima = df['IDADE'].max()
print(f"\nIdade Média dos Alunos: {idade_media:.2f}")
print(f"Idade Mínima dos Alunos: {idade_minima}")
print(f"Idade Máxima dos Alunos: {idade_maxima}")

# 1.4 Contar a quantidade de alunos por turno (MANHÃ, TARDE, NOITE, INTEGRAL).
contagem_turno = df['TURNO'].value_counts()
print("\nContagem de Alunos por Turno:")
print(contagem_turno)

# 1.5 Contar a quantidade de alunos por modalidade de ensino.
contagem_modalidade = df['MODALIDADE'].value_counts()
print("\nContagem de Alunos por Modalidade de Ensino:")
print(contagem_modalidade)

# 1.6 Identificar qual turma possui o maior número de alunos.
contagem_turma = df['TURMA'].value_counts()
turma_mais_alunos = contagem_turma.index[0]
num_alunos_turma = contagem_turma.iloc[0]
print(f"\nTurma com o maior número de alunos: '{turma_mais_alunos}' com {num_alunos_turma} alunos.")
```

## 2. Visualização de Dados (Gráficos)

```python
# 2.1 Gráfico de barras: Quantidade de alunos por sexo.
plt.figure(figsize=(8, 5))
sns.countplot(x='SEXO', data=df, palette='viridis')
plt.title('Quantidade de Alunos por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade de Alunos')
plt.show()
```

```python
# 2.2 Gráfico de barras: Quantidade de alunos por turno.
plt.figure(figsize=(10, 6))
sns.countplot(x='TURNO', data=df, palette='magma', order=df['TURNO'].value_counts().index)
plt.title('Quantidade de Alunos por Turno')
plt.xlabel('Turno')
plt.ylabel('Quantidade de Alunos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2.3 Histograma: Distribuição das idades dos alunos.
plt.figure(figsize=(10, 6))
sns.histplot(df['IDADE'], bins=15, kde=True, color='skyblue')
plt.title('Distribuição das Idades dos Alunos')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# 2.4 Boxplot: Distribuição de idade por sexo.
plt.figure(figsize=(10, 6))
sns.boxplot(x='SEXO', y='IDADE', data=df, palette='pastel')
plt.title('Distribuição de Idade por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.show()
```

```python
# 2.5 Gráfico de barras: Top 10 turmas com mais alunos.
top_10_turmas = df['TURMA'].value_counts().head(10)
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_turmas.index, y=top_10_turmas.values, palette='cubehelix')
plt.title('Top 10 Turmas com Mais Alunos')
plt.xlabel('Turma')
plt.ylabel('Quantidade de Alunos')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2.6 Gráfico de barras: Idade média por turno.
idade_media_por_turno = df.groupby('TURNO')['IDADE'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=idade_media_por_turno.index, y=idade_media_por_turno.values, palette='viridis')
plt.title('Idade Média por Turno')
plt.xlabel('Turno')
plt.ylabel('Idade Média')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
